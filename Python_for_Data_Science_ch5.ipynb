### 5.6 Parametric  (linear) correlation analysis
- A method you can use to find correlation between linearly related continuous numeric variables

- Pearson Correlation Coefficient
  - Assumption: data is norally distributed; continuous numeric variables; variables are linearly related
  - Use Pearson Correlation:
      - To uncover (linear) relationships between variables
      - Not to rule out possible (nonlinear) relationships

  * R = 1 -> Strong positive relationship
  * R = 0 -> Not linearly correlated
  * R = -1 -> Strong negative relationship[linktekst](https://)

### 5.6 Non-parametric (non-linear) correlation analysis

- to find correlation between categorical, nonlinearly related, non-normally distributed variables
- Spearman's rank correlation
  * Find the R correlation between variable-pairs of ordinal data type (ordinal variables are bumeric and able to be categorized)
  * Variable-pairs are then able to be ranked according to the strength of the correlation between them
  * R=1, R=0 & R=-1 the same as Pearson's rank
  * Assumptions: variables are ordinal, numeric, nonlinearly, non-normally distributed
- Chi-square Test for Independence
  * p <0.05 -> reject null hypothesis and variables are correlated
  * p > 0.05 -> accept null hypothesis and conclude that the variables are independent

  * Use Chi-square test if:
      - variables are categorical or numeric
      - variables are binned (variables could be broken up into 10 seperated groups)
      

```python
# create a small subset data pairs
## all of these variables are ordinal, nonlinear, nonnormal

X = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(X)
```

```python
import scipy
from scipy.stats import spearmanr
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

#### Chi-square test for independence

```python
# create a cross table
table = pd.crosstab(cyl, am)

# import Chi-square test

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)

print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, vs)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, gear)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

### 5.7 Transforming dataset distribution

- Aim of scaling data
  * so that different magnitude among variables do not produce erroneous or misleading statistics
  * to prepare data for machine learning

- Two ways to scale data
  * Normalization: putting each observation on a relative scale between the values of 0 and 1:
  
        value of observation/sum of all observation in variable

  * Standardization: rescaling data so that it has a zero mean and unit variance

- Use Python scikit-learn library to preproces: scaling, centering, normalizing, binning and imputing data

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
# data visualization settings

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
address = 'C:/Users/Lillian/Desktop/ExerciseFiles/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
# to generate some summary statistics on this variable

cars[['mpg']].describe()
```

```python
# to make sure that a one-column matrix will be generated
mpg_matrix = mpg.values.reshape(-1,1)

# To transform our variable by scaling to a defined range (by default it's between 0 and 1)
scaled = preprocessing.MinMaxScaler()

# to fit the min/max scaler transformer to our data and return a transformed version of data
scaled_mpg = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg)
```

```python
# To define exact range of your variable to be scaled to
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
# Using scale() to scale your features
# by defaut this function will center the data and scale it to unit variance
standardized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
plt.plot(standardized_mpg)
# no mean, no standard diviation
```

```python
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg)

# the mean has been set to 0, the distribution has unit variance
# the variable has a standard normal distribution and it has been standardized
```

### 5.8 Extreme value analysis for outliers

- Outlier detection
  * Most machine learning methods assume that your data has been treated for outliers
  * Detecting outliers can be a data preprocessing task or an analytical method of its own merit

- Use outlier detection to uncover anomalies that represent:
  * Equipment Failure
  * Fraud
  * Cybersecurity Event

- Tukey methods = Tukey Boxplots
  * Inter-quartile rang (IQR): or spread the distance between the first quartile (at 25%) and the third quartile (at 75%)

- Univariate Method: Tukey Outlier Labeling
  - a = Q1 - 1.5(IQR)
  - b = Q3 + 1.5(IQR)
     If ...
  - min.value < a, or
  - max.value > b, then ...
  The variable is suspect for ooutliers
