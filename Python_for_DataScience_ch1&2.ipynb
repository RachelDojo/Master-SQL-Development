## 2.2 Treating missing Values in Python
- By default, missing values are represented with NaN:"Not a Number"
- Warning: If your dataset has 0s, 99s or 999s, be sure to either drop or approximate them as you would with missing values
- Way 1: take the average missing values in order to generate an approximation

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
##Figuring out what data is missing

missing = np.nan
series_obj = Series(['row1', 'row2', missing, 'row4', 'row5', 'row6', missing, 'row8'])
series_obj
```

```python
#isnull method: runs a boolean value that describes in true or false whether an element in a Pandas object is a null value (null value is missing value)

series_obj.isnull()
```

```python
#How to fill in for missing values
## create a data frame
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
##with .loc method to select rows and columns and then set certain values equal to missing (in the ['row value', 'column value])

DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

```python
## With .fillna() method to fill in the missing values with a numeric value that you specify
# in this case fill in missing value with 0

filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
#fill in the missing values from each column series as designated by the dictionary key with its own value as specified by the corresponding dictionary value{}

filled_DF = DF_obj.fillna({0:0.1, 5:1.25})
filled_DF
```

How this could be useful?
- predictive application needs 4 variables as input, one variable has missing values. To fill in an approximation in order to make the predictive app work
- pass in the method 'ffill' argument  

```python
# 'ffill' method: fill in missing values with values from the last non-null element in the column series

fill_DF = DF_obj.fillna(method='ffill')
fill_DF

```
